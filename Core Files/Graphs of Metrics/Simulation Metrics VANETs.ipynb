```python
import numpy as np
import matplotlib.pyplot as plt

percentages = np.array([10, 20, 30, 40])

# Data for Highway
accuracy_highway = np.array([0.881142, 0.659081, 0.530829, 0.562283]) * 100
precision_highway = np.array([0.900343, 0.567024, 0.438477, 0.581667]) * 100
recall_highway = np.array([0.451587, 0.169944, 0.110955, 0.183105]) * 100
specificity_highway = np.array([0.99195, 0.933636, 0.902557, 0.90201]) * 100
f1_highway = np.array([0.544821, 0.379887, 0.290143, 0.374896]) * 100
avg_detection_time_highway = np.array([20.359, 8.60988, 10.5883, 17.4182])

# Data for City
accuracy_city = np.array([0.834817, 0.712713, 0.661115, 0.548668]) * 100
precision_city = np.array([0.63498, 0.481974, 0.500629, 0.645758]) * 100
recall_city = np.array([0.492628, 0.333438, 0.344177, 0.294826]) * 100
specificity_city = np.array([0.923527, 0.905618, 0.870747, 0.838722]) * 100
f1_city = np.array([0.601813, 0.449925, 0.489654, 0.446683]) * 100
avg_detection_time_city = np.array([29.0313, 25.2332, 18.6973, 43.949])

# Adjust Y-axis limits and ticks
y_ticks_time = np.arange(0, max(max(avg_detection_time_highway), max(avg_detection_time_city)) + 10, 10)
y_ticks_metrics = np.arange(0, 101, 10)

# Plotting Highway data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(percentages, accuracy_highway, label='Accuracy')
plt.plot(percentages, precision_highway, label='Precision')
plt.plot(percentages, recall_highway, label='Recall')
plt.plot(percentages, specificity_highway, label='Specificity')
plt.plot(percentages, f1_highway, label='F1 Score')
plt.yticks(y_ticks_metrics)
plt.xlabel('Malicious Nodes in Network (percentile)')
plt.ylabel('Metrics (percentile)')
plt.title('Highway Metrics')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(percentages, avg_detection_time_highway, label='Trust Model', color='red')
plt.yticks(y_ticks_time)
plt.xlabel('Malicious Nodes in Network (percentile)')
plt.ylabel('Time (Epochs)')
plt.title('Highway Average Ghost Detection Time')

plt.tight_layout()
plt.show()

# Plotting City data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(percentages, accuracy_city, label='Accuracy')
plt.plot(percentages, precision_city, label='Precision')
plt.plot(percentages, recall_city, label='Recall')
plt.plot(percentages, specificity_city, label='Specificity')
plt.plot(percentages, f1_city, label='F1 Score')
plt.yticks(y_ticks_metrics)
plt.xlabel('Malicious Nodes in Network (percentile)')
plt.ylabel('Metrics (percentile)')
plt.title('City Metrics')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(percentages, avg_detection_time_city, label='Trust Model', color='red')
plt.yticks(y_ticks_time)
plt.xlabel('Malicious Nodes in Network (percentile)')
plt.ylabel('Time (Epochs)')
plt.title('City Average Ghost Detection Time')

plt.tight_layout()
plt.show()
```
